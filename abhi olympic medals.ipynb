```python
#Bycity

def f(x):
    return Series(dict(events_unique = x['Edition'].nunique()
                                             
                       ))

                 
k = data.groupby(data['City']).apply(f)

k = k.sort_values(by='events_unique', ascending=False)

k.plot.bar()
```

```python
#By edition


def f(x):
    return Series(dict(Atheletes_count = x['Athlete'].nunique()
                                             
                       ))

                 
k = data.groupby(data['Edition']).apply(f)



k.plot.bar()

```

```python
#sports&events

def f(x):
    return Series(dict(sport_count = x['Sport'].nunique(),
                       Event_count = x['Event'].nunique()
                                             
                       ))

                 
k = data.groupby(data['Edition']).apply(f)



k.plot(kind='line',secondary_y='Event_count')
```

```python
#Participatingcountries

def f(x):
    return Series(dict(country_count = x['NOC'].nunique()
                       
                                             
                       ))

                 
k = data.groupby(data['Edition']).apply(f)



k.plot(kind='bar')

```

```python
#Women%

k = data[['Edition','Athlete','Gender']].drop_duplicates()

def f(x):
    return Series(dict(Men_count = x['Gender'].apply(lambda x: (x=='Men')).sum(),
                       Women_count = x['Gender'].apply(lambda x: (x=='Women')).sum()
                                             
                       ))


k = k.groupby('Edition').apply(f)

k['Women%'] = (k['Women_count']/(k['Women_count']+k['Men_count']))*100

k

k[['Women%']].plot(kind='bar')
```

```python
#Medalpercentage


def f(x):
    return Series(dict(Gold_count = x['Medal'].apply(lambda x: (x=='Gold')).sum(),
                       Silver_count = x['Medal'].apply(lambda x: (x=='Silver')).sum(),
                       Bronze_count =  x['Medal'].apply(lambda x: (x=='Bronze')).sum()                                          
                       ))


k = data.groupby('Edition').apply(f)
k['Gold%'] = (k['Gold_count']/(k['Gold_count']+k['Silver_count']+k['Bronze_count']))
k['Silver%'] = (k['Silver_count']/(k['Gold_count']+k['Silver_count']+k['Bronze_count']))
k['Bronze%'] = (k['Bronze_count']/(k['Gold_count']+k['Silver_count']+k['Bronze_count']))


k[['Gold%','Silver%','Bronze%']].plot(kind='bar',stacked='true')

```

```python
#overall medals -avg
def f(x):
    return Series(dict(Medal_count = x['Medal'].count()))

                 
k = data.groupby(['Edition','NOC']).apply(f)

k = k.groupby(['NOC']).mean()


k = k.sort_values('Medal_count',ascending=False).head(10)

k.plot(kind='bar')


```
